### 4. 保單的「初始認列日」
##### 保險合約群組(group)應於下列日期中最早者認列：
- 該群組保障期間開始日
- 該群組之被保險人第一期保費到期日
- 對於虧損性合約群組，當該群組已成為虧損時點

```python
Q1_data[ '初始認列日' ] = np.nan

# 選取合約群組中最早認列的日期
first_date = Q1_data[ [ '保費到期日', '保險起日' ] ].min( 1 ).min( )

# 判斷群組中每張保單的「初始認列日」
for i in np.arange( len( Q1_data[ '保批單號碼' ] ) ) :
    if min( Q1_data[ '保費到期日' ][i], Q1_data[ '保險起日' ][i] ) == first_date :
        Q1_data[ '初始認列日' ][i] = first_date
    elif  min( Q1_data[ '保費到期日' ][i], Q1_data[ '保險起日' ][i] ) > first_date :
        Q1_data[ '初始認列日' ][i] = min( Q1_data[ '簽單日' ][i], Q1_data[ '保費到期日' ][i], Q1_data[ '保險起日' ][i] ) 
    else :
        Q1_data[ '初始認列日' ][i] = np.nan
        
# Q1_data
```

### 5. 計算未到期負債(LRC)

##### 5-1 計算「保期(月)」及「當期已經過月數」

#### 輸入評估日期

```python
""" 輸入評估日 """
valuation_date = '2018-03-31'
```

```python
# 定義計算函數

def months( data, evaluate_date ) :
    # 計算保期(月份數)
    start_year = data[ '保險起日' ].apply( lambda x : x.year )
    end_year = data[ '保險迄日' ].apply( lambda x : x.year )
    start_month = data[ '保險起日' ].apply( lambda x : x.month )
    end_month = data[ '保險迄日' ].apply( lambda x : x.month )
    period = ( end_year - start_year )*12 + ( end_month - start_month + 1 )
    data[ '保期(月)' ] = period
    
    # 計算當期已經過月數
    evaluate_month = datetime.strptime( evaluate_date, '%Y-%m-%d' ).date( ).month
    data[ '當期已經過月數' ] =  evaluate_month - start_month + 1
    
    return data
```

```python
months( Q1_data,  valuation_date )  # 新增欄位：保期(月)、當期已經過月數
```

##### 5-2 計算「期初LRC」

```python
Q1_data[ '期初LRC' ] = np.nan  # 創造新的空白欄位

for i in np.arange( len( Q1_data[ '保批單號碼' ] ) ) :   
    if Q1_data[ '當期已經過月數' ][i] <= 3 :
        Q1_data[ '期初LRC' ][i] = 0
    else : 
        Q1_data[ '期初LRC' ][i] = 1

Q1_data
```

##### 5-3 計算「當期已收保費」、「當期已付佣金」
- 當期已收保費 = 保費收取金額
- 當期已付佣金 = 佣金支付金額
- 對於當期未收取保費的保單，將「保費收取金額」及「佣金支付金額」補 0 

```python
Q1_data = Q1_data.fillna( { '保費收取金額' : 0, '佣金支付金額' : 0 } )   
```

##### 5-4 計算「當期認列之合約收入」、「取得成本攤銷」
- $ \displaystyle 當期認列之合約收入= 簽單保費  \times \frac{當期已經過月數}{保期(月)} $  
######    
- $ \displaystyle 取得成本攤銷 = 簽單保費  \times 佣金率 \times \frac{當期已經過月數}{保期(月)} $

```python
Q1_data[ '當期認列之合約收入' ] =  Q1_data.apply( lambda X : X[ '簽單保費' ] * ( X[ '當期已經過月數' ] / X[ '保期(月)' ] ), axis = 1 )
Q1_data[ '取得成本攤銷' ] = Q1_data.apply( lambda X : X[ '簽單保費' ] * X[ '佣金率' ] * ( X[ '當期已經過月數' ] / X[ '保期(月)' ] ), axis = 1 )
Q1_data
```

##### 5-5 計算「期末LRC」
- $ 期末LRC = 期初LRC + 當期已收保費 - 當期已付佣金 - 當期認列之合約收入 + 取得成本攤銷 $

```python
Q1_data[ '期末LRC' ] =  Q1_data.apply( lambda X : X[ '期初LRC' ] + X[ '保費收取金額' ] - X[ '佣金支付金額' ] - X[ '當期認列之合約收入' ] + X[ '取得成本攤銷' ] , axis=1 )
Q1_data
```

##### 5-5 計算「累積已收保費」、「累積已付佣金」

```python
Q1_data[ '累積已收保費' ] = Q1_data[ '保費收取金額' ] 
Q1_data[ '累積已付佣金' ] = Q1_data[ '佣金支付金額' ] 
Q1_data
```

### 6. 財務報導與揭露
- 資產負債表 Statement of Financial Position
- 綜合損益表 Statement of Financial Performance

```python
select_col = [ '保批單號碼', '保費收取金額', '佣金支付金額', '期末LRC', '當期認列之合約收入', '取得成本攤銷' ]
Q1_statement_data = Q1_data[ select_col ]
Q1_statement_data
```

#### 資產負債表 Statement of Financial Position

```python
Position_table = pd.DataFrame( columns = [ '現金', '合約資產', '合約負債' ], index = [ '2018Q1_amounts' ]  )

# 當期現金 = 前期現金 + 當期已收保費 - 當期已付佣金
Position_table.loc[ '2018Q1_amounts', '現金']  = Q1_statement_data[ '保費收取金額' ].sum( ) - Q1_statement_data[ '佣金支付金額' ].sum( ) 

# 合約資產、合約負債
LRC = Q1_statement_data[ '期末LRC' ].sum( )
if LRC > 0 :
    Position_table.loc[ '2018Q1_amounts', '合約負債' ] = LRC
    Position_table.loc[ '2018Q1_amounts', '合約資產' ] = 0
elif LRC < 0 :
    Position_table.loc[ '2018Q1_amounts', '合約資產' ] = LRC
    Position_table.loc[ '2018Q1_amounts', '合約負債' ] = 0
else : 
    Position_table.loc[ '2018Q1_amounts', '合約資產' ] = 0 
    Position_table.loc[ '2018Q1_amounts', '合約負債' ] = 0

# 業主權益 = 前期現金 + 合約資產 - 合約負債
Position_table[ '業主權益' ] = Position_table.apply( lambda X : X[ '現金' ] + X[ '合約資產' ] - X[ '合約負債' ], axis = 1 )

print( '2018Q1 財務狀況表' )
Position_table
```

#### 綜合損益表 Statement of Financial Performance

```python
Performance_table = pd.DataFrame( columns = [ '保險合約收入', '已發生賠款', '取得成本攤銷',  ], index = [ '2018Q1_amounts' ]  )

Performance_table.loc[ '2018Q1_amounts', '保險合約收入' ] = Q1_statement_data[ '當期認列之合約收入' ].sum( )
Performance_table.loc[ '2018Q1_amounts', '取得成本攤銷' ] = Q1_statement_data[ '取得成本攤銷' ].sum( )
Performance_table.loc[ '2018Q1_amounts', '已發生賠款' ] = 0
Performance_table[ '保險損益' ] = Performance_table.apply( lambda X : X[ '保險合約收入' ] - X[ '已發生賠款' ] - X[ '取得成本攤銷' ], axis=1 )

print( '2018Q1 財務績效表' )
Performance_table
```
